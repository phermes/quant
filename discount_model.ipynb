```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 2.80
n   = range(1,20)
```

```python
def stockvalue(eps,nmax=20,interest=1.01):
    EPSvals = [ eps/((interest)**i) for i in range(1,nmax)]
    return sum(EPSvals)
```

```python
stockvalue(2.08)
```
