```python
from bs4 import BeautifulSoup
import urllib2
import json
import pandas as pd
import os.path
```

## Morningstar API

```python
rawurl = "http://financials.morningstar.com/ajax/ReportProcess4CSV.html?t={}&reportType={}&period=12&dataType=A&order=asc&columnYear=10&number=3"
```

```python
def get_statement(ticker,statetype,cname):
    '''Possible types of statement: is, bs, cf'''
    
    if statetype not in ['is','bs','cf']:
        raise ValueError("Possible types of statement: is, bs, cf")
    
    direct       = {}
    direct['is'] = 'XETRA_ms_income_statement/'
    direct['bs'] = 'XETRA_ms_balance/'
    direct['cf'] = 'XETRA_ms_cashflow/'
    
    url     = rawurl.format(ticker,statetype)
    freq    = urllib2.Request(url, headers={'User-Agent' : "Firefox"}) 
    con     = urllib2.urlopen( freq ) # content of the web file
    html    = con.read()
    soup    = BeautifulSoup(html,"html.parser")
    
    html = soup.prettify("utf-8")
    with open("{}{}-{}.csv".format(direct[statetype], ticker, cname), "wb") as file:
        file.write(html)
```

```python
def get_keyratios(ticker,cname):
    
    rawurl  = "http://financials.morningstar.com/ajax/exportKR2CSV.html?t={}"    
    direct  = 'XETRA_ms_keyratios/'
    
    url     = rawurl.format(ticker)
    freq    = urllib2.Request(url, headers={'User-Agent' : "Firefox"}) 
    con     = urllib2.urlopen( freq ) # content of the web file
    html    = con.read()
    soup    = BeautifulSoup(html,"html.parser")
    
    html = soup.prettify("utf-8")
    with open("{}{}-{}.csv".format(direct, ticker,cname), "wb") as file:
        file.write(html)
```
